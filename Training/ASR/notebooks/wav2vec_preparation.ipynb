Now creating vocabulary from these labels

```python
def extract_all_chars(df):
    list_of_labels = df["labels"]
    chars = set()
    for label in list_of_labels:
        for char in label:
            chars.add(char)
    return list(chars)
```

```python
vocab_list = extract_all_chars(df1)
vocab_list
```

```python
pd.set_option('display.max_columns', None)  # or 1000
pd.set_option('display.max_rows', None)  # or 1000
pd.set_option('display.max_colwidth', None)  # or 199
```

```python
def addCountAndTotal(df):
    print(f'There are total {df.shape[0]} rows in the dataset')
    print('Iteration started')
    for i in range(df.shape[0]):
        for c in df.iloc[i]['labels']:
            # if c == ' ' or c == '\u200c' or c == '\u200d': c = 'space'
            if c not in vocab_list: continue
            df.iloc[i,df.columns.get_loc(c)] += 1
        if i % 10000 == 0:
            print(f'{i} rows done')
    print('Iteration Completed.\nNow Calculating total')
    df['total'] = df.iloc[:,2:].sum(axis=1) 
    print('Total Calculated\nProcess Completed!!')
    return df 
```

```python
df1 = addCountAndTotal(df1)
df1.head()
```

```python
def get_details(filepath):
    data,sr = librosa.load(filepath)
    n_samples = data.shape[0]
    sampling_rate = sr 
    duration = librosa.get_duration(y=data,sr=sr)
    return n_samples,sampling_rate,duration,data
```

```python
def checkLessThan5sec(filepath):
    data,sr = librosa.load(filepath)
    duration = librosa.get_duration(y=data,sr=sr)
    return duration <= 5
```

```python
df1['n_samples'] = 0
df1['sampling_rate'] = 0
df1['duration'] = 0 
df1.head()
```

```python
def add_details(df):
    print(f'=> This dataframe has {df.shape[0]} files')
    print('=> Processing Started')
    for i in range(df.shape[0]):
        filepath = f"D:\Programming\Projects\major_project\Codes\ASR\data_old\\audio\\{df['path'][i]}.flac"
        n_samples,sampling_rate,duration = get_details(filepath)
        df['n_samples'][i] = n_samples
        df['sampling_rate'][i] = sampling_rate
        df['duration'][i] = duration

        if i!= 0 and i%1000==0:
            print(f'=> Completed {i} files')
            df.to_csv('D:\Programming\Projects\major_project\Codes\ASR\csv_checkpoints\\temp_data_{i}.csv',index=False)
    print('=> Processing Complete')
    return df 
```
