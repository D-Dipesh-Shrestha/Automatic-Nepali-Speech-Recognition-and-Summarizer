Finding duration, no of samples for each data

```python
df3 = pd.read_csv('../input/asr-cnn-dataset/new_transcript_with_char_count.csv')
df3.head()
```

```python
import librosa
from IPython.display import Audio
```

```python
def get_details(filepath):
    data,sr = librosa.load(filepath)
    n_samples = data.shape[0]
    sampling_rate = sr 
    duration = librosa.get_duration(y=data,sr=sr)
    return n_samples,sampling_rate,duration
```

```python
df3['n_samples'] = 0
df3['sampling_rate'] = 0
df3['duration'] = 0 
df3.head()
```

Function to add details to each file

```python
def add_details(df):
    print(f'=> This dataframe has {df.shape[0]} files')
    print('=> Processing Started')
    for i in range(df.shape[0]):
        filepath = f"../input/asr-cnn-dataset/audio/audio/{df.loc[i,'filename']}.flac"
        n_samples,sampling_rate,duration = get_details(filepath)
        df.loc[i,'n_samples'] = n_samples
        df.loc[i,'sampling_rate'] = sampling_rate
        df.loc[i,'duration'] = duration

        if i!= 0 and i%1000==0:
            print(f'=> Completed {i} files')
    print('=> Processing Complete')
    return df 
```
