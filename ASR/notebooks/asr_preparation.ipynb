```python
df3 = pd.read_csv('D:\Programming\Projects\major_project\Codes\ASR\data_old\\transcript\\new_dataset1.csv')
df3.drop(['Unnamed: 0'],axis=1,inplace=True)
df3.head()
```

```python
unique_chars = set() 
for label in df3['label'].to_list():
    for char in label:
        unique_chars.add(char)

print(len(unique_chars))
unique_chars
```

Finding count of each character to delete unwanted characters

```python
for char in unique_chars:
    df3[char] = 0 
df3.head()
```

Calculate total count of characters in each label

```python
def addCountAndTotal(df):
    print(f'There are total {df.shape[0]} rows in the dataset')
    print('Iteration started')
    for i in range(df.shape[0]):
        for c in df.iloc[i]['label']:
            # if c == ' ' or c == '\u200c' or c == '\u200d': c = 'space'
            if c not in unique_chars: continue
            df.iloc[i,df.columns.get_loc(c)] += 1
        if i % 10000 == 0:
            print(f'{i} rows done')
    print('Iteration Completed.\nNow Calculating total')
    df['total'] = df.iloc[:,2:].sum(axis=1) 
    print('Total Calculated\nProcess Completed!!')
    return df 
```

```python
df3 = addCountAndTotal(df3)
df3.head()
```

```python
char_count = {} 
for char in unique_chars:
    char_count[char] = df3[char].sum()
sorted_char_count = sorted(char_count.items(),key=lambda x:x[1], reverse=True)
sorted_char_count
```

To remove the characters whose count is less than 100

```python
chars_to_remove = {char:char_count[char] for char in unique_chars if char_count[char] <= 100}
chars_to_remove_l = list(chars_to_remove.keys())
chars_to_remove_l
```

To check if a character is present in label or not

```python
def containsChar(label,char):
    return char in label
```

Assign check = 1 if label contains unwanted characters and then drop those labels

```python
df3['check'] = 0 
for i in range(df3.shape[0]):
    for char in chars_to_remove_l:
        if containsChar(df3['label'][i],char):
            df3['check'][i] = 1
            break 
print(df3.shape)
df3.head()
```

```python
df3 = df3[df3['check']==0]
print(df3.shape)
df3.head()
```

drop the check column

```python
df3.drop(columns='check',inplace=True)
print(df3.shape)
df3.head()
```

Sort the values by total

```python
df3 = df3.sort_values(by="total",ascending=False)
df3.head()
```

Finding duration, no of samples for each data

```python
import librosa
from IPython.display import Audio
```

```python
def get_details(filepath):
    data,sr = librosa.load(filepath)
    n_samples = data.shape[0]
    sampling_rate = sr 
    duration = librosa.get_duration(y=data,sr=sr)
    return n_samples,sampling_rate,duration
```

```python
df3['n_samples'] = 0
df3['sampling_rate'] = 0
df3['duration'] = 0 
df3.head()
```

Function to add details to each file

```python
def add_details(df):
    print(f'=> This dataframe has {df.shape[0]} files')
    print('=> Processing Started')
    for i in range(df.shape[0]):
        filepath = df['path'][i]
        n_samples,sampling_rate,duration = get_details(filepath)
        df['n_samples'][i] = n_samples
        df['sampling_rate'][i] = sampling_rate
        df['duration'][i] = duration

        if i!= 0 and i%1000==0:
            print(f'=> Completed {i} files')
    print('=> Processing Complete')
    return df 
```
